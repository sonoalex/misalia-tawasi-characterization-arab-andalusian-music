```python
import json
import pprint

result = {}

with open("andalusian_description.json") as jsonFile:
    andalusian_json_objects = json.load(jsonFile)
    
        
    for andalusian_object in andalusian_json_objects:
        if len(andalusian_object['sections']) > 0:
            result.update({andalusian_object['mbid']: andalusian_object['sections'][0]['nawba']['transliterated_name']})
        
        else:
            result.update({andalusian_object['mbid']: 'none (No sections)'})


pprint.pprint(result)

with open('mbid_nawba.json', 'w') as jsonFile:
    json.dump(result, jsonFile)
    jsonFile.close()
```
